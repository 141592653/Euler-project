Solution :

 $$3\sum_{k=0}^{\lfloor \frac{n-1}{3} \rfloor} k + 5\sum_{k=0}^{\lfloor \frac{n-1}{5} \rfloor} k - 15\sum_{k=0}^{\lfloor \frac{n-1}{15} \rfloor} k $$

 $$ = \frac{1}{2} \Big(3\lfloor \frac{n-1}{3} \rfloor(\lfloor \frac{n-1}{3} \rfloor + 1) + 5\lfloor \frac{n-1}{5} \rfloor(\lfloor \frac{n-1}{5} \rfloor + 1) - 15\lfloor \frac{n-1}{15} \rfloor(\lfloor \frac{n-1}{15} \rfloor + 1)\Big)$$

```python
n = 1000
```

```python
floor3  = (n - 1) // 3
floor5  = (n - 1) // 5
floor15 = (n - 1) // 15
```

```python
res = 3 * floor3 * (floor3 + 1) + 5 * floor5 * (floor5 + 1) - 15 * floor15 * (floor15 + 1)
res //= 2
```
